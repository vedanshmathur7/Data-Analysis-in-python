 # Plot a simple line graph of y=x^2 for values of x from -10 to 10 using matplotlib. Add labels, titles and grid.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x ** 2
plt.plot(x, y, label='y = x²', color='green')
plt.title('Plot of y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Draw a histogram of random normal distribution values using both matplotlib and seaborn. Compare the results

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1, 1000 samples

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram (Matplotlib)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.histplot(data, bins=30, kde=True, color='salmon')
plt.title('Histogram with KDE (Seaborn)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

# Plot multiple lines on the same graph: y=x,y=x*2,y=x*3. Use legends, different line styles, and markers.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 200)
y1 = x
y2 = x**2
y3 = x**3

plt.plot(x, y1, label='y = x', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='y = x²', color='green', linestyle='--', marker='s')
plt.plot(x, y3, label='y = x³', color='red', linestyle='-.', marker='^')

# Labels, title, legend, grid
plt.title('Multiple Line Plot: y = x, x², x³')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

# Plot a stacked bar chart showing the population of 3 countries (India, USA, China)over years 2000,2010,2020 using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
years = ['2000', '2010', '2020']
x = np.arange(len(years))
```

```python
india_pop = [1.05, 1.21, 1.38]
usa_pop = [0.28, 0.31, 0.33]
china_pop = [1.26, 1.34, 1.41]

# Create stacked bar chart
plt.figure(figsize=(10, 6))

# Plot China at bottom
plt.bar(x, china_pop, label='China', color='red')
# Stack India on top of China
plt.bar(x, india_pop, bottom=china_pop, label='India', color='orange')
# Stack USA on top of India + China
total_bottom = np.array(china_pop) + np.array(india_pop)
plt.bar(x, usa_pop, bottom=total_bottom, label='USA', color='blue')
plt.title('Population of China, India, and USA (2000–2020)')
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.xticks(x, years)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# **Simulate 3 stock prices using random walk and plot them on single graph with different styles

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters
days = 100
stocks = 3
start_price = [100, 150, 200]

# Simulate random walk
stock_prices = []
for i in range(stocks):
    returns = np.random.normal(loc=0.001, scale=0.02, size=days)
    price = start_price[i] * np.cumprod(1 + returns)
    stock_prices.append(price)

plt.plot(stock_prices[0], label='Stock A', linestyle='-', color='blue')
plt.plot(stock_prices[1], label='Stock B', linestyle='--', color='green')
plt.plot(stock_prices[2], label='Stock C', linestyle='-.', color='red')

plt.title('Simulated Stock Prices using Random Walk')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

#**Using seaborn's tips dataset, create a facetgrid of scatter plots (total_bill vs tip) separated by time

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Create a FacetGrid separated by 'time'
g = sns.FacetGrid(tips, col='time')

# Map scatter plots to the grid (total_bill vs tip)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip')

g.set_axis_labels('Total Bill', 'Tip')
g.set_titles(col_template="{col_name} Time")
plt.show()
```
